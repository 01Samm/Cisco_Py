```python
 # Operadores aritméticos


a = 10
b = 3


print(a +  b)     # 13      (adição)
print(a -  b)     # 7       (subtração)
print(a *  b)     # 30      (multiplicação)
print(a /  b)     # 3.3333  (divisão)
print(a // b)     # 3       (divisão inteira)
print(a %  b)     # 1       (módulo/resto da divisão)
print(a ** b)     # 1000    (exponenciação)
print(a ** (1/b)) # 2.14    (radiciação)
```

```python
# Operadores relacionais ou de comparação


a = 10
b = 3


print(a == b)  # False  Igual
print(a != b)  # True   Diferente
print(a >  b)  # True   Maior
print(a <  b)  # False  Menor
print(a >= b)  # True   Maior ou igual
print(a <= b)  # False  Menor ou igual
```

```python
# Operadores lógicos


a = True
b = False


print(a and b)  # False
print(a or b)   # True
print(not a)    # False
```

```python
# FATIAMENTO DE STRINGS


texto = "Curso de Python"


print(texto[6])          # 7º caracterer
print(texto[9:15])       # 10º ao 15º caracterer
print(texto[9:15:2])     # 10º ao 15º caracterer de 2 em 2
print(texto[:5])         # Até o 5º caracterer
print(texto[9:])         # 10º caracterer em diante
print(texto[::-1])       # Inverter o texto
print(texto[-1])         # Último caracter
```

```python
# ANÁLISE DE STRINGS (Case sensitive)


frase = "SENAI Ary Torres"


# tamanho da string (quantidade de caracteres)
print(len(frase))


# contar caracteres (ou sequência de caracteres)
print(frase.count('e'))


# encontrar índice (retorna -1 se não encontrar)
print(frase.find('I'))


# encontrar índice (a partir da direita)
print(frase.rfind('s'))


# encontrar índice (gera erro se não encontrar)
print(frase.index('r'))


# verificar pertinência
print('ENA' in frase)
```

```python
# OPÇÃO ÚNICA


operacao = input("""
ESCOLHA UMA OPÇÃO:


[1] Opção 1
[2] Opção 2
[3] Opção 3
""")


match operacao:
    case '1':
        print("Opção 1 foi escolhida...")
    case '2':
        print("Opção 2 foi escolhida...")
    case '3':
        print("Opção 3 foi escolhida...")
    case _:
        print("Opção inválida...")
```

```python
dia = int(input("""
ESCOLHA UM DIA DA SEMANA:

[1] Segunda-feira
[2] Terça-feira
[3] Quarta-feira
[4] Quinta-feira
[5] Sexta-feira
[6] Sábado
[7] Domingo

"""))

match dia:

    case 1:
        print("\nSegunda-feira")
    case 2:
        print("\nTerça-feira")
    case 3:
        print("\nQuarta-feira")
    case 4:
        print("\nQuinta-feira")
    case 5:
        print("\nSexta-feira")
    case 6:
        print("\nSábado")
    case _:
        print("\nDomingo")
```

```python
letra = input("Digite uma letra: ").upper()

match letra:

    case "A" | "E" | "I" | "O" | "U":
        print("\nVocê digitou uma vogal")
    case _ if letra.isalpha():
        print("\nVocê digitou uma consoante")
    case _ if letra.isnumeric():
        print("\nVocê digitou um número inteiro")
    case _:
        print("\nVocê digitou um caractere qualquer")
```

```python
""" A diferença entre while e for em Python está principalmente em como e quando eles controlam a repetição. Vou te explicar de forma simples e prática 👇

🔁 while → repete enquanto uma condição for verdadeira

Use quando você não sabe exatamente quantas vezes vai repetir.

Exemplo:
contador = 1

while contador <= 5:
    print(contador)
    contador += 1
🧠 Como funciona:
Ele verifica a condição (contador <= 5)
Se for True, executa
Repete até a condição virar False

👉 Muito usado quando:

Depende de entrada do usuário
Depende de alguma condição mudar
🔄 for → percorre uma sequência (lista, range, etc.)

Use quando você já sabe quantas vezes quer repetir ou tem uma coleção de dados.

Exemplo:
for i in range(1, 6):
    print(i)
🧠 Como funciona:
Ele percorre os valores de range(1, 6)
Executa uma vez para cada valor

👉 Muito usado quando:

Você tem listas, strings, arrays
Quer repetir um número fixo de vezes
⚔️ Comparação direta
Característica	while	for
Controle	Condição (True/False)	Sequência (lista, range, etc.)
Quando usar	Não sei quantas repetições	Sei quantas repetições
Risco	Loop infinito ⚠️	Mais seguro
Exemplo típico	Entrada de usuário	Percorrer lista
💡 Exemplo comparando os dois
Com while:
i = 0
while i < 3:
    print(i)
    i += 1
Com for:
for i in range(3):
    print(i)

👉 Ambos fazem a mesma coisa, mas o for é mais simples nesse caso.

🚀 Resumão direto
while → repete até a condição parar de ser verdadeira
for → repete um número definido de vezes ou percorre algo

"""
```

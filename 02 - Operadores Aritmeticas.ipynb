```python
#exercicio do professor de produto sendo descontado

#Desconto: informe o valor de um produto e o percentual de desconto.
#Calcule o valor do desconto e o preço final. Exiba os resultados.

produto = float(input("Digite o valor do produto: "))
desconto = float(input("Digite o valor do desconto: "))
VD = produto*(desconto/100)
print(f"O valor do produto com desconto é: {produto - VD}")
```

```python
#Exercicio 3
#Velocidade Média: Informe a distância percorrida e o tempo gasto.
#Calcule a velocidade média usando a fórmula: Vm = Distância / Tempo

distancia = float(input("Digite a distancia percorrida: "))
tempo = float(input("Digite o tempo gasto: "))
vm = distancia / tempo

print(f"A velocidade média é: {vm:.1f} km/h")
```

```python
# Exercício 2:
# Desconto: Informe o valor de um produto e o percentual de desconto.
# Calcule o valor do desconto e o preço final. Exiba os resultados.


valor = float(input('Digite o valor do produto: '))
desconto = float(input('Digite a % de desconto: '))


vrdesconto = valor * (desconto/100) # 10/100 - 50/100 - 1000 * 0.90
vrfinal = valor-vrdesconto


print(f'Valor do produto: R$ {valor:.2f}')
print(f'% de desconto: {desconto}%')
print(f'Valor do desconto: R$ {vrdesconto:.2f}')
print(f'Valor final: R$ {vrfinal:.2f}')
```

```python
# Exercício 1:
# Conversor de Temperatura: temperatura em Celsius. Ex: 25ºc
# Converta para Fahrenheit usando a fórmula: (0°C * 9/5) + 32 = 32°F
# Imprima a temperatura em Fahrenheit.


temperatura = float(input('Digite a temperatura em ºC: '))
fanrenheit = temperatura*(9/5)+32


print(f'A temperatura de {temperatura}ºC em Fahrenheit é de {fanrenheit:.1f}ºF.')
```

```python
#Exercicio 4
#Volume do Cubo:
#Solicite a aresta do cubo.
#Calcule o volume usando a fórmula: volume = aresta³

aresta = float(input("informe a aresta do cubo: "))
print(f"O volume do cubo é: {aresta**3}")
```

```python
#Exercicio 5
#Solicite o ano de nascimento.
#Calcule a idade considerando o ano atual
#Exiba a idade.

nascimento = int(input("Digite o ano de nascimento: "))
ano_atual = int(input("Digite o ano atual: "))
idade = ano_atual - nascimento
print(f"Sua idade é: {idade}")
```

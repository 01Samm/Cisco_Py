```python
# Crie um programa que faça o computador jogar Jokenpô com você:

from random import choice
from google.colab import output
from time import sleep

vitoria_atom = 0
vitoria_zeus = 0
empates = 0
contador = 0

opcao = ['Pedra','Papel','Tesoura']

while contador < 500000:
    atom = choice(opcao)
    zeus = choice(opcao)

    if zeus == atom:
        # print('Empate 🤝')
        empates += 1
    elif (zeus == 'Tesoura' and atom == 'Papel') or (zeus == 'Papel' and atom == 'Pedra') or (zeus == 'Pedra' and atom == 'Tesoura'):
        # print('Zeus venceu!!! 🍾🏆')
        # print(f'Zeus --> {zeus}  x {atom} <-- Atom')
        vitoria_zeus += 1
    elif (atom == 'Tesoura' and zeus == 'Papel') or (atom == 'Papel' and zeus == 'Pedra') or (atom == 'Pedra' and zeus == 'Tesoura'):
        # print('Atom venceu!!! 🤖🎉')
        # print(f'Zeus --> {zeus} x {atom} <-- Atom')
        vitoria_atom += 1
    else:
        print('Jogada Inválida!')
    contador += 1

print('\n\nPlacares:')
print(f'Zeus: {vitoria_zeus}')
print(f'Atom: {vitoria_atom}')
print(f'Empate: {empates}')
```

```python
#Acerte a letra

import string
from random import choice
from time import sleep
from google.colab import output

alfabeto_minusculo = list(string.ascii_lowercase)

letra_secreta = choice(alfabeto_minusculo)

while True:
    letra_escolhida = input("Escolha uma letra: ").lower()

    if letra_escolhida == letra_secreta:
        print("Você acertou!!!")
        break
    elif letra_escolhida != letra_secreta:
        print("Você errou essa é maior!!!")
    elif letra_escolhida != letra_secreta:
        print("Você errou essa é menor!!!")
    else:
        print("Invalido")
        sleep(3)
        output.clear()
    print('Fim do programa!!')
```

```python
# Exercicio:
# Desenvolver um programa que vai solicitar diversos número (positivos/negativos)
# e após o comando de saída ele vai exibir qual o maior
# número e qual o menor número digitado, bem como a quantidade
# de números digitados.
```

```python
quantidade = 0
maior = 0
menor = 0

while True:
    numero = input("Digite um numero qualquer ("s" para sair): ")
    if numero == "s":
        break
    else:
        numero = float(numero)
        quantidade += 1
        if numero > maior:
            maior = numero
        if numero < menor:
            menor = numero
print(f"\nNumeros digitados: {quantidade}")
print(f"Maior numero: {maior}")
print(f"Menor numero: {menor}")
```

```python
# Exercício:
# Desenvolver um programa que vai solicitar diversos números (positivos/negativos)
# e após o comando de saída ele vai exibir qual o maior
# número e qual o menor número digitado, bem como a quantidade
# de números digitados.


quantidade = 0
maior = 0
menor = 0


while True:
    numero = input('Digite um número qualquer ("S" para sair: ').lower()


    if numero == 's':
        break
    else:
        numero = float(numero)
        quantidade += 1


        if numero > maior:
            maior = numero
        if numero < menor:
            menor = numero


print(f'\nNúmeros digitados: {quantidade}')
print(f'O maior número foi: {maior}')
print(f'O menor número foi: {menor}')

```

```python
# faça um programa que escolha um número aleatório entre 1 e 500
# tente adivinhar o número; o programa te dará dicas se é maior ou menor


import random
from time import sleep
from google.colab import output


numero_secreto = random.randint(1, 500)


while True:
    numero = int(input('Adivinhe o número (1 a 500): \n'))


    if numero == numero_secreto:
        print('Você acertou o número 🥳🎉!')
        break
    elif numero > numero_secreto:
        print('Seu número é maior! ⬆️')
    elif numero < numero_secreto:
        print('Seu número é menor! ⬇️')
    else:
        print('Entrada inválida')
    sleep(3)
    output.clear()
```

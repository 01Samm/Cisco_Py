```python
# Faça um programa que leia o peso de 12 pessoas.
# No final, mostre qual foi o maior e o menor peso lidos.
```

```python
maior_peso = 0
menor_peso = float('inf') # Inicializa com um valor muito grande

for i in range(1, 13):
    peso = float(input(f"Digite o peso da pessoa {i} (em kg): "))
    if peso > maior_peso:
        maior_peso = peso
    if peso < menor_peso:
        menor_peso = peso

print(f"O maior peso lido foi: {maior_peso} kg")
print(f"O menor peso lido foi: {menor_peso} kg")
```

```python
# #Desenvolva um programa que leia dez números inteiros e mostre a soma apenas
# daqueles que forem pares. Se o valor digitado for ímpar desconsidere-o.
# contar os números pares digitados.
```

```python
par = 0
contador = 0

for i in range(1, 11):
    num = int(input(f"Digite o {i}º número: "))
    if num % 2 == 0:
        par += num
        contador += 1

print(f"A soma dos números pares é: {par}")
print(f"Quantidade de numeros pares: {contador}")
```

```python
# Faça um programa que peça a idade de 15 pessoas
# mostre quantas são maiores de idade e quantas são menores de idade
```

```python
maior = 0
menor = 0

for i in range(1, 16):
    idade = int(input(f"digite a idade da pessoa {i}: "))
    if idade >= 18:
        maior += 1
    else:
        menor += 1

print(f"Total de pessoas maiores de idade: {maior}")
print(f"Total de pessoas menores de idade: {menor}")
```

```python
# Faça um programa que peça o ano de nascimento de 15 pessoas
# mostre quantas são maiores de idade e quantas são menores de idade
from time import sleep


maiores = 0
menores = 0
ano_atual = 2026


for i in range(1,16):
    ano = int(input(f'Digite o ano de nascimento da pessoa {i}: '))
    idade = ano_atual - ano
    if idade >= 18:
        maiores += 1
    else:
        menores += 1
    sleep(1)
    print(f'Idade: {idade}')


print(f'Quantidade de maiores: {maiores}')
print(f'Quantidade de menores: {menores}')
```

```python
# recriar o exercício Maior Número / Menor Número usando listas


numeros = []


while True:
    num = input('Digite um número ("S" para Sair): ').lower()


    if num == 's':
        break
    else:
        num = float(num)
        numeros.append(num)
media = sum(numeros)/len(numeros)
print(f'Maior: {max(numeros)}')
print(f'Menor: {min(numeros)}')
print(f'Soma:  {sum(numeros)}')
print(f'Quant: {len(numeros)}')
print(f'Média: {media}')
```

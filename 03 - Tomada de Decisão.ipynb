```python
# Exercício : Aprovação de Empréstimo
# Objetivo: Verificar aprovação de empréstimo com novos critérios:
# Valor do empréstimo (R$)
# Prazo (meses)
# Cálculo Automático: Valor da parcela = (valor do empréstimo / prazo)
# Requisitos para aprovação:
# Renda mínima: R$ 3.000
# Score de crédito: >= 500
# Parcela máxima: <= 35% da renda
# Número de parcelas: Máx 36


renda = float(input('Digite o valor da sua renda: '))
score = float(input('Digite o valor do score informado pelo SERASA: '))


if renda >= 3000 and score >= 500:
    renda35 = renda * (35/100)
    print(f'Você pode comprometer até R$ {renda35:.2f} da sua renda.')

    vr_emprestimo = float(input('Informe o valor do empréstimo: '))
    qt_parcelas = int(input('Informe a quantidade de parcelas: (Máx:36) '))

    if qt_parcelas <= 36:
        vr_parcela = vr_emprestimo / qt_parcelas


        if vr_parcela <= renda35:
            print(f'Empréstimo aprovado em {qt_parcelas}x de R$ {vr_parcela:.2f} ')
        else:
            print(f'Valor da parcela maior do que o valor permitido! ')
            print(f'O valor da parcela escolhida foi R$ {vr_parcela:.2f}')
    else:
        print('Número de parcelas superior ao permitido! ')
else:
    print('Empréstimo negado!')
```
